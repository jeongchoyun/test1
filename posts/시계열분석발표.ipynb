# 시계열분석 발표

### 데이터

- 2020.10.08 ~ 2021.12.31 대한민국 누적 코로나 확진자 수
- 2021.11월 이후 누적 확진자 수가 대폭 증가하는 경향. (델타 + 오미크론 영향)

```r
options(repr.plot.width = 12, repr.plot.height = 8)
data  = read.csv("COVID.csv")
z = data$n;head(z)
```

```r
forecast::tsdisplay(z, lag.max = 28, main = "COVID-19 time series")
```

- 시도표 상 이분산성이 보임. 또한 ACF 그림을 보아 비정상성 시계열 및 주기가 7인 계절성분이 보이는 것 같음.
- 주기가 7인 이유: 같은 주 기준 평일 월, 화에 높고 금,토,일 주말에는 낮은 경향을 보였음.

- 따라서 분산 안정화, 차분, 계절 차분할 예정.

```r
forecast::BoxCox.lambda(z)
```

```r
boxcox_z = forecast::BoxCox(z,lambda= forecast::BoxCox.lambda(z));head(boxcox_z)
```

```r
forecast::tsdisplay(boxcox_z, lag.max = 28, main = "Boxcox data")
```

```r
# 단위근검정 H0 : phi=1
fUnitRoots::adfTest(boxcox_z, lags = 1, type = "ct")
fUnitRoots::adfTest(boxcox_z, lags = 2, type = "ct")
```

```r
d_boxcox_z = diff(boxcox_z)
forecast::tsdisplay(d_boxcox_z, lag.max = 28, main = "differencing data")
```

```r
ds_d_boxcox_z = diff(d_boxcox_z, 7)

forecast::tsdisplay(ds_d_boxcox_z, lag.max = 28, main = "seasonal differencing data")
```

- 비계절형 ACF에서 1,2 시차에서 절단/ PACF에서는 1,2,3,4? 에서 절단
- 비계절형에서는 차수가 더 낮은 ARIMA(0,0,2)모형 선택
- 계절형 ACF 7에서 절단, PACF는 7,14,21에서 절단 
- 계절형에서는 SARIMA(0,0,1)7
- 따라서 ds_d_boxcox_z = ARIMA(0,0,2)(0,0,1)7 or boxcox_z = ARIMA(0,1,2)(0,1,1)7을 선택

### 모형 적합

```r
fit1 = arima(ds_d_boxcox_z, order = c(0,0,2),
             seasonal = list(order = c(0,0,1), period=7))
summary(fit1)
```

```r
lmtest::coeftest(fit1)
```

```r
fit2 = arima(boxcox_z, order = c(0,1,1),
             seasonal = list(order = c(0,1,1), period=7))
summary(fit2)
```

```r
lmtest::coeftest(fit2)
```

## 잔차검정

```r
forecast::checkresiduals(fit2, main = "residual")
```

```r
t.test(resid(fit2))
```

```r
Box.test(resid(fit2), lag=1, type = "Ljung-Box")
Box.test(resid(fit2), lag=6, type = "Ljung-Box")
Box.test(resid(fit2), lag=12, type = "Ljung-Box")
```

```r
astsa::sarima(boxcox_z, 0,1,1,0,1,1,7, details = T)
```

### 여러가지 모형 적합

```r
summary(fit2)$aic
summary(fit2)$sigma2
```

```r
fit3 = arima(boxcox_z, order = c(1,1,0),
             seasonal = list(order = c(0,1,1), period=7))
summary(fit3)
lmtest::coeftest(fit3)
summary(fit3)$aic
summary(fit3)$sigma2
```

```r
fit4 = arima(boxcox_z, order = c(2,1,0),
             seasonal = list(order = c(0,1,1), period=7))
summary(fit4)
lmtest::coeftest(fit4)
summary(fit4)$aic
summary(fit4)$sigma2
```

```r
fit5 = arima(boxcox_z, order = c(0,1,1),
             seasonal = list(order = c(0,1,2), period=7))
summary(fit5)
lmtest::coeftest(fit5)
summary(fit5)$aic
summary(fit5)$sigma2
```

```r
fit6 = arima(boxcox_z, order = c(1,1,1),
             seasonal = list(order = c(1,1,1), period=7))
summary(fit6)
lmtest::coeftest(fit6)
summary(fit6)$aic
summary(fit6)$sigma2
```

```r
fit7 = forecast::auto.arima(ts(boxcox_z, frequency=7),
                     test = "adf",
                     seasonal = TRUE, trace = F)
summary(fit7)
lmtest::coeftest(fit7)
summary(fit7)$aic
summary(fit7)$sigma2
```

- AIC, Sigma^2 가장 작은 ARIMA(2,1,0)(0,1,1)7 적합

### 예측

```r
fore_fit = forecast::forecast(fit2, 31)

forecast_z = forecast::InvBoxCox(fore_fit$mean, lambda=forecast::BoxCox.lambda(z))

forecast_z_upper95 = forecast::InvBoxCox(fore_fit$upper[,2], lambda=forecast::BoxCox.lambda(z))
forecast_z_lower95 = forecast::InvBoxCox(fore_fit$lower[,2], lambda=forecast::BoxCox.lambda(z))
```

```r
plot(1:(length(z)+31), c(z, forecast_z_upper95), type='n', xlab='t', ylab='z', main = "Prediction for 1 month ahead")
lines(1:length(z),z)
lines((length(z)+1) : (length(z)+31), forecast_z, col='red')
lines((length(z)+1) : (length(z)+31), forecast::InvBoxCox(fore_fit$lower[,2], lambda=forecast::BoxCox.lambda(z)), col='blue')
lines((length(z)+1) : (length(z)+31), forecast::InvBoxCox(fore_fit$upper[,2], lambda=forecast::BoxCox.lambda(z)), col='blue')
```
