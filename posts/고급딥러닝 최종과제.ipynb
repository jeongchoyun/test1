### 2) DNN 모델 적합

```python
df3 = df2.dropna()
```

```python
df3_dummies = pd.get_dummies(df3[["SEX", "REGION", "INCOME", "GLY_CD", "OLIG_OCCU_CD", "OLIG_PROTE_CD", "FAMILIY_DM", "SMOKE" , "ALCOHOL", "RPA"]])
df3_dummies = df3_dummies.astype(int)
df3_dummies.head()
```

```python
df3_dummies_columns = ["SEX", "REGION", "INCOME", "GLY_CD", "OLIG_OCCU_CD", "OLIG_PROTE_CD", "FAMILIY_DM", "SMOKE" , "ALCOHOL", "RPA"]
```

```python
df3_dummies_not_columns = list(set(df3.columns) - set(df3_dummies_columns));df3_dummies_not_columns
df3_not_dummies = df3[df3_dummies_not_columns];df3_not_dummies.head()
```

```python
df4 = pd.concat([df3_dummies, df3_not_dummies], axis = 1);df4.head()
```

```python
train_df_DM_YES = df4[df4['isDM'] == 1]

train_df_DM_NO = df4[df4['isDM'] == 0].sample(n= len(train_df_DM_YES))

train_df_DNN = pd.concat([train_df_DM_YES, train_df_DM_NO])
```

```python
train_DNN = train_df_DNN.sample(frac=0.8,random_state=202350543)

test_DNN = train_df_DNN.drop(train_DNN.index)

train_DNN_stats = train_DNN.describe()

train_DNN_stats.pop("isDM")

train_DNN_stats = train_DNN_stats.transpose()
```

```python
train_labels = train_DNN.pop('isDM')

test_labels = test_DNN.pop('isDM')

def norm(x):
  return (x - train_DNN_stats['mean']) / train_DNN_stats['std']

normed_train_DNN = norm(train_DNN)

normed_test_DNN= norm(test_DNN)
```

```python
def build_model():
  model = keras.Sequential([

    layers.Dense(128, activation='relu', input_shape=[len(normed_train_DNN.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')])

  optimizer = tf.keras.optimizers.Adam(0.0001)
  model.compile(loss='binary_crossentropy',  # 분류 문제에 적합한 손실 함수 선택
                  optimizer=optimizer,
                  metrics=['accuracy'])

  return model
```

```python
model = build_model()
model.summary()

class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

EPOCHS = 500
history = model.fit(
  normed_train_DNN, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[early_stop, PrintDot()])
# callbacks=[early_stopping],
```

```python
pd.DataFrame(history.history)
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.plot(hist['epoch'], hist['loss'], label = 'Train loss')
    
  plt.plot(hist['epoch'], hist['val_loss'], label = 'Validation loss')
  plt.title('Loss')
  plt.legend()
 
    
  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.plot(hist['epoch'], hist['accuracy'], label = 'Train accuracy')

  plt.plot(hist['epoch'],hist ['val_accuracy'], label = 'Validation val_accuracy')

  plt.title('Accuracy')
  plt.legend()
  plt.show()

plot_history(history)
```

`-` train, test 셋 accuracy, loss

```python
model.evaluate(normed_train_DNN, train_labels, verbose=2)
```

```python
model.evaluate(normed_test_DNN, test_labels, verbose=2)
```

```python
threshold = 0.5
pred_y = np.where(model.predict(normed_test_DNN) > threshold, 1, 0)
```
